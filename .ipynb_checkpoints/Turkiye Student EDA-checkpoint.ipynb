# Turkiye Student Evaluation
In this project, we aims to explore Turkiye student dataset. The dataset itself can be found on Kaggle or UCI Machine Learning repository. We begin by importing modules we would probably need later.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

We now import the dataset

```python
# Get and read the data
direct_link = 'http://archive.ics.uci.edu/ml/machine-learning-databases/00262/turkiye-student-evaluation_generic.csv'
df = pd.read_csv(direct_link)
```

We make sure that the dataset contains no missing data

```python
df.isnull().values.any()
```

Alright, seems that we are good to go

## Explore the Data

We begin by asking which class has the most active student. The student enrolled in that class will give the most response to the class. One problem with this is that the classes don't necessarily to have same amount of student. But, for now, we assume that they do (Bold assumption, I know :p)

```python
df['class'].value_counts()
```

Turns out class 3 has the most response (with class 13 closely following). Let's graph it to see it visually.

```python
# Plot frequency of responses vs class
cc = sns.countplot(x='class', data=df)
cc.set(ylabel='number of responses')
```

Okay, we know that class 3 has the most response. Lets talk about that response. For now, we dont know whether, based on the response, the students liked or disliked the class. In other word, we want to see good how class 3 is among students.

Let's filter all the class 3 first.

```python
# filter the dataset to data with response from class 3 only
df_3 = df[df['class'] == 3]
```

Course 3 is the has the most response, which means it has the most student. It is reasonable to assume that course 3 is probably first year course-an introductory course, so it must be relatively easy for students. Let's see if that is true.

```python
df_3['difficulty'].value_counts()
```

... and we are indeed correct. 334 students say it is level 1 difficult.
One thing that is interesting is that 204 said it is level 4 difficult and only 28 said it is level 28 difficult. That's weird. How could level 4 is the 2nd most responses. I would assume that level 2 should've been the 2nd most responses. Why is there such a drastic changes (level 1 and level 4)??

Anyway, let's graph it.

```python
cd = sns.countplot(x='difficulty', data=df_3)
cd.set(ylabel='number of responses')
plt.title('Difficulty of class 3')
```

It is safe to say that the lower the difficulty is, the higher likeness of student against the class is. One big portion student like the class while another big portion dislike it (hence a lot of 3s, 4s and 5s). Let's see if the student who say it is easy really like it or not.

We filter the data and get data where student say class 3 is easy.

```python
# filter student who say class 3 is easy
df_3_easy = df_3[df_3['difficulty'] == 1]
```

We are going to examine the following quesonnaire response:
- Q3: The course was worth the amount of credit assigned to it.
- Q8: The quizzes, assignments, projects and exams contributed to helping the learning.
- Q9: I greatly enjoyed the class and was eager to actively participate during the lectures.

I think those questions are good measures of how liked the course 3 is.

```python
df_3_easy_ques = df_3_easy[['Q3', 'Q8', 'Q9']]
qc = sns.countplot(x="value", hue="variable", data=pd.melt(df_3_easy_ques))
plt.title('Response against course 3 from Q3, Q8, Q9')
plt.legend(loc='center left', bbox_to_anchor=(1, .5), ncol=1, title='Questions: ')
```

As we can see, the response is pretty similar among each question. Turns out, most student says course 3 is average in all of these questions. Also, a big part of them dislike he the course (we can see a huge number of 1s given as response)

We can then conclude:
Student who said the class being easy (gave level 1 difficulty) didn't necessarily like the class (not much 5s from Q3, Q8, Q9), in fact they said the course is mostly average (based on Q3, Q8 and Q9)

Could it be instructur issue? Concidentally, course 3 is managed by instructor 3. Let's find out.

### How is instructor 3 doing?
Lets first filter the data and get all response in instructor 3's course.

```python
# get response data from all course of instructor 3
df_instr_3 = df[df['instr'] == 3]
```

```python
df_instr_3['class'].value_counts()
```

Instructor 3 manage 7 courses, lets examine how the instrutor is doing with the top 5 of those courses.

```python
# get the top 5 courses
top_5 = [3, 13, 5, 9, 8]

# filter the data by those top 5 courses
df_instr_3_top5 = df_instr_3[df_instr_3['class'].isin(top_5)]
```

Let's look how the instructor is doing on the following questions:
- Q13: The Instructor's knowledge was relevant and up to date.
- Q20: The Instructor explained the course and was eager to be helpful to students
- Q23: The Instructor encouraged participation in the course.

All of these questions are related to how the instructor is doing on the course

```python
df_instr_3_top5_ques = df_instr_3_top5[['Q13', 'Q20', 'Q23']]
```

```python
ic = sns.countplot(x="value", hue="variable", data=pd.melt(df_instr_3_top5_ques))
plt.title('Response against instructor 3 from Q13, Q20, Q23')
plt.legend(loc='center left', bbox_to_anchor=(1, .5), ncol=1, title='Questions: ')
```

Mostly average. However, we can see a lot of 4s and 5s, which mean the instructor is generally well-liked and doing good in the course.

#### How about instructor 3 as a person? How well is he/her doing?

Let's see how instructor 3 doing personally. We first get all response from instructor 3's class.

We are going to examine the instructor on the following questions:
- Q14: The Instructor came prepared for classes
- Q17: The Instructor arrived on time for classes.
- Q19: The Instructor made effective use of class hours.
- Q21: The Instructor demonstrated a positive approach to students.
- Q28: The Instructor treated all students in a right and objective manner.

```python
# lets take all those questions column name to a list
personal_ques = ['Q14', 'Q17', 'Q19', 'Q21', 'Q28']

df_instr_3_personal = df_instr_3[personal_ques]
prsc = sns.countplot(x="value", hue="variable", data=pd.melt(df_instr_3_personal))
plt.title('Response against instructor 3 from personal question')
plt.legend(loc='center left', bbox_to_anchor=(1, .5), ncol=1, title='Questions: ')
```

Conclusion: instructor 3 is a good instrutor. He/she came prepared, arrived on time, used office hours, was positive against student and treated student fairly. Instructor 3 is just the ideal instructor you could think of :p

- The previous problem is certainly not because of the instructor. Instructor 3 (who also manage course 3) is doing just fine in his/her class. The instructor is well-liked among students.
